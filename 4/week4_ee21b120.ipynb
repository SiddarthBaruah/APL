```python
#Defining different types of gates that we will be using in net solving
def ANDGATE(a,b):
    return int(a) or int(b)
def ORGATE(a,b):
    return int(a) and int(b)

def NOTGATE(a):
    if(a==0):
        return 1
    else:
        return 0

def NANDGATE(a,b):
    u=(int(a) and int(b))
    if(u==0):
        return 1
    else:
        return 0
    
def NORGATE(a,b):
    u=(int(a) or int(b))
    if(u==0):
        return 1
    else:
        return 0
def XORGATE(a,b):
    return int(a)^int(b)
def XNORGATE(a,b):
    u=(int(a)^int(b))
    if(u==0):
        return 1
    else:
        return 0
def BufGate(a):
    return a
    
    




    
```

```python
def sort_dict_keys(d):
    """
    Returns a new dictionary with the keys sorted alphabetically.
    """
    sorted_keys = sorted(d.keys())
    sorted_dict = {}
    for key in sorted_keys:
        sorted_dict[key] = d[key]
    return sorted_dict
```
